```python
from langchain_chroma import Chroma
from agent_blocks import embeddings_model
```

```python
vector_store = Chroma(
    collection_name="academic-data",
    embedding_function=embeddings_model,
    persist_directory="./academic-data"
)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
file_path = '/Users/a/Programming/Langchain-Project/notebooks-langchain-before-project/external-data/telegram-chat.txt'
```

```python
loader = TextLoader(file_path)
```

```python
documents = loader.load()
```

```python
assert len(documents)
print(f"Total character: {len(documents[0].page_content)}")
```

```python
print(documents[0].page_content[:600])
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500, #chunk size (characters)
    chunk_overlap=50, # overlap (Characters)
    add_start_index=True,
)
```

```python
all_splits = text_splitter.split_documents(documents)
```

```python
print(f"split blog post into {len(all_splits)} sub-documents.")
```

```python
print(all_splits[40])
```

```python
document_ids = vector_store.add_documents(documents=all_splits)
```

```python
test = vector_store.similarity_search("Time: Oct 10, 2025 08:00 Jakarta", k=3)
```

```python
for doc in test:
    print(f"{doc.page_content}\n-----")
```
