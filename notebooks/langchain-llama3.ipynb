````python
import requests
import json
````

````python
# url model llama3:8b local
url = 'http://localhost:11434/api/chat'
````

````python
def llama_generate(prompt):
    data = {
        "model":"llama3:8b",
        "messages": [{
            "role":"user",
            "content": prompt
        }
        ],
        "stream":False
    }
    response = requests.post(url, json=data)
    return response.json()['message']['content']
````

````python
llama_generate('tanggal berapa hari ini')
````

### Implementasi langchain chat template untuk boost model LLM

````python
# implementasi akan menggunakan contoh mengganti style tulisan atau email
````

````python
template_string = """
    terjemahkan text dalam triple backtick kedalam style dalam bentuk {style} \
    text:  ```{text}```
"""
````

````python
customer_style = """
    bahasa indonesia yang lebih sopan
"""
````

````python
costumer_email = """
Arrr, aku sedang marah besar, tutup blender-ku
terbang dan memercikkan dinding dapurku
dengan smoothie! Dan yang lebih parah lagi,
garansinya tak menanggung biaya
buat bersihin dapurku. Aku butuh bantuanmu
sekarang juga, sobat bajak laut!
"""
````

````python
# check input variable pada chat-template
chat_template = ChatPromptTemplate.from_template(template_string)

# variable input pada chat_template
chat_template.messages[0].prompt.input_variables
````

````python
customer_messages = chat_template.format_messages(
    style=customer_style,
    text=costumer_email
)
````
